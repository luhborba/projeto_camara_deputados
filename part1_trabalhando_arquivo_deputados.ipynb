```python
# Importações de Bibliotecas Necessárias
import pandas as pd
```

```python
# Criando Data Frame
df_deputados = pd.read_csv('deputados.csv', delimiter=";")
df_deputados.head()
```

```python
# Verificando Informações do DF
df_deputados.info()
```

```python
# Verificando Tamanho do DF
df_deputados.shape
```

```python
#Removendo o url através da função replace com Regex, criando assim a coluna ID
df_deputados['id'] = df_deputados['uri'].replace('.*/(\\d+)', r'\1', regex=True)
df_deputados.head()
```

```python
# Verificando duplicados na coluna id
duplicados = df_deputados['id'].duplicated()
display(df_deputados[duplicados])
```

```python
# Verificando atualmente em exercício
df_deputados_2023 = df_deputados.query('idLegislaturaInicial == 57 or idLegislaturaFinal == 57')
df_deputados_2023
```

```python
# Verificando tamanho do novo DataFrame
df_deputados_2023.info()
```

```python
# Eliminando Colunas que não desejo usar
df_deputados_2023 = df_deputados_2023.drop(columns=['uri',
                                                    'urlWebsite',
                                                    'cpf',
                                                    'urlRedeSocial',
                                                    'dataFalecimento'
                                                   ])
# Reorganizando ordem das colubas
df_deputados_2023 = df_deputados_2023[['id', 'nome','nomeCivil','siglaSexo','dataNascimento','ufNascimento','municipioNascimento','idLegislaturaInicial','idLegislaturaFinal']]
```
